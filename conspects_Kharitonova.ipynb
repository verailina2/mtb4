## 3 отличия от классической математики

### Python позволяет выполнять разные математические вычисления, но он имеет некоторые отличия от классической математики. 

### 1. Python использует числа с плавающей точкой для представления действительных чисел, в то время как классическая математика использует десятичные дроби или несократимые дроби. Стоит уделять внимание тому, какие типы деления и округления используются в коде, так как это может привести к неточностям в вычислениях в Python.

### Типы округления:

int() Округляет число в сторону нуля

```python
int(4.65678)
```

round(x)  Округляет число x до ближайшего целого. Если дробная часть числа равна 0.5, то число округляется до ближайшего четного числа

```python
round(3.987)
```

round(x, n) Округляет число x до n знаков после точки

```python
round(5.98765, 3)
```

floor(x)  Округляет число x вниз («пол»)

```python
import math
math.floor(7.45678)
```

ceil(x) Округляет число x вверх («потолок»)

```python
import math
math.ceil(7.45678)
```

abs(x)  Модуль числа x (абсолютная величина)

```python
abs(7.45678)
```

### Типы деления (см. примеры выше):


* Деление (/)
* Целочисленное деление (//)
* Остаток от деления (%)

```python
import math
math.sqrt(9)
```

```python
import math
math.pow(9, 2)
```

```python
import math
math.degrees(1)
```

```python
import math
math.cos(1)
```

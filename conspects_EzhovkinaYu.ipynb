##1. Работа с комплексными числами. 
Комплексные числа можно складывать, вычитать, умножать, делить и возводить в степень. Для создания комплексного числа можно использовать функцию complex(a,b), в которую, в качестве первого аргумента, передается действительная часть, а в качестве второго - минимальная. Либо записать как a + bj.

```python
z = 1 + 2j
print (z)
```

```python
x = complex(3, 2)
print (x)
```

```python
x + z
```
